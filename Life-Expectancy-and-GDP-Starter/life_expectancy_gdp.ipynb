```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading data and displaying head
df = pd.read_csv('all_data.csv')
df.head()
```

```python
#understanding data
df.info()
```

```python
#change United States of America to USA. Change the name of a life expentancy column
df.Country = df.Country.replace('United States of America', 'USA')
df = df.rename(columns={'Life expectancy at birth (years)':'expectency'})
```

```python
big_GDP = df[(df.Country=='USA') | (df.Country=='China')]
medium_GDP = df[(df.Country=='Germany') | (df.Country=='Mexico')]
small_GDP = df[(df.Country=='Chile') | (df.Country=='Zimbabwe')]
```

```python
###Bivariate analysis

#H0 - there is no difference between a country and its GDP
#H1 - there is difference between a country and its GDP

sns.boxplot(x='Country', y='GDP', data=df)
plt.title('The GDP per country')
plt.show()
plt.clf()
```

```python
#lets look at life expentancy and country

sns.boxplot(x='Country', y='expectency', data=df)
plt.show()
```

```python
#line of all countries GDP

sns.lineplot(x='Year', y='GDP', hue='Country', data=df)
plt.legend(loc='upper left')
plt.show()
plt.clf()
```

```python
#line of life expectency and year

sns.lineplot(x='Year', y='expectency', hue='Country', data=df)
plt.legend(loc='upper left')
plt.show()
plt.clf()
```

```python
#overall GDP and life expectency
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=df)
plt.title('GDP and Life Expectency yearly')
plt.show()
plt.clf()
```

```python
#scatter plot of GDP and life expentency 

sns.scatterplot(x='expectency', y='GDP', hue='Country', data=df)
#plt.legend(loc='upper left')
plt.show()
plt.clf()
```

```python
#Lets investigate the change of GDP in Zimbabwe and the life expectency:
zim_df = df[df.Country=='Zimbabwe']
USA_df = df[df.Country=='USA']
Ger_df = df[df.Country=='Germany']
Mex_df = df[df.Country=='Mexico']
Chn_df = df[df.Country=='China']
Chi_df = df[df.Country=='Chile']

plt.figure(figsize=(15, 15))

plt.subplot(2, 3, 1)
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=USA_df)
plt.title('USA GDP and Life Expectency yearly')

plt.subplot(2, 3, 2)
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=Chn_df)
plt.title('China GDP and Life Expectency yearly')

plt.subplot(2, 3, 3)
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=Ger_df)
plt.title('Germany GDP and Life Expectency yearly')

plt.subplot(2, 3, 4)
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=Mex_df)
plt.title('Mexico GDP and Life Expectency yearly')

plt.subplot(2, 3, 5)
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=Chi_df)
plt.title('Chile GDP and Life Expectency yearly')

plt.subplot(2, 3, 6)
sns.scatterplot(x='expectency', y='GDP', hue='Year',data=zim_df)
plt.title('Zimbabwe GDP and Life Expectency yearly')

plt.subplots_adjust(wspace=1, hspace=0.4)

plt.show()
plt.clf()
```

```python
#Overall outcome: we can see that growing GDP relates to growing life expectency. However, there is not much data to find 
#a causal relationship. We would need further data to investigate the reasons.
```
